```python
class Model(torch.nn.Module):
    def __init__(self):
        super().__init__()
        w = 0.5
        self.w = torch.nn.Parameter(torch.full((1,), w, dtype = torch.float32))
                                             
        pass#def 
    
    def forward(self, x):
        x = self.w*x
        return x
        pass#def 
    pass#class
model = Model()
loss_fn = torch.nn.L1Loss()
opt = torch.optim.RMSprop(model.parameters(), lr = 1e-4)

X = torch.tensor([0.1], dtype = torch.float32)
Y = torch.tensor([0.1], dtype = torch.float32)

for epoch in range(10):
    opt.zero_grad()
    pred = model(X)
    loss = loss_fn(pred, Y)
    loss.backward()
    opt.step()
    
    print(model.w.data)
    pass

```

```python
class Model(torch.nn.Module):
    def __init__(self, order):
        super().__init__()
        w = 0.5
        self.order = order
        w_like = math.pow(w, 1/self.order)
        self.w_like = torch.nn.Parameter(torch.full((1,), w_like, dtype = torch.float32))
                                             
        pass#def 
    
    def forward(self, x):
        x = self.get_w()*x
        return x
        pass#def 
    def get_w(self):
        return torch.pow(self.w_like, self.order)
    pass#class
models = [Model(order) for order in [1,2,3,5,10,20]]
loss_fn = torch.nn.MSELoss()


X = torch.tensor([0.1], dtype = torch.float32)
Y = torch.tensor([0.1], dtype = torch.float32)

for model in models:
    print(colored(F"{model.order}---------------","green"))
    opt = torch.optim.RMSprop(model.parameters(), lr = 1e-3)
    
    for epoch in range(1000):
        opt.zero_grad()
        pred = model(X)
        loss = loss_fn(pred, Y)
        if loss.item()<1e-8:
            print(epoch)
            break
            pass
        loss.backward()
        opt.step()
        #print(model.w_like)
        pass    
    w = model.get_w()
    print(w.data)
    pass
```

```python
class Model(torch.nn.Module):
    def __init__(self, order):
        super().__init__()
        w = 1
        b = 0.1
        self.order = order
        w_like = math.pow(w, 1/self.order)
        b_like = math.pow(b, 1/self.order)
        self.w_like = torch.nn.Parameter(torch.full((1,), w_like, dtype = torch.float32))
        self.b_like = torch.nn.Parameter(torch.full((1,), b_like, dtype = torch.float32))
                                             
        pass#def 
    
    def forward(self, x):
        x = self.get_w()*x + self.get_b()
        return x
        pass#def 
    def get_w(self):
        return torch.pow(self.w_like, self.order)
        pass#def 
    def get_b(self):
        return torch.pow(self.b_like, self.order)
        pass#def
    pass#class
models = [Model(order) for order in [1,-0.5,2,2.263, -1,3,4,4.34, -3, 5,10,15, 20, 0.5, 0.1,   ]]
loss_fn = torch.nn.MSELoss()


X = torch.tensor([1], dtype = torch.float32)
Y = torch.tensor([3], dtype = torch.float32)

for model in models:
    print(colored(F"{model.order}---------------","green"))
    opt = torch.optim.RMSprop(model.parameters(), lr = 1e-3)
    
    for epoch in range(2000):
        opt.zero_grad()
        pred = model(X)
        loss = loss_fn(pred, Y)
        if loss.item()<1e-8:
            print(epoch)
            break
            pass
        loss.backward()
        opt.step()
        #print(model.w_like)
        pass    
    w = model.get_w()
    print(w.data)
    b = model.get_b()
    print(b.data)
    pass
```
